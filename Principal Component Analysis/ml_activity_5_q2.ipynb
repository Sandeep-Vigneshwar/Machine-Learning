##PCA for housing-related dataset - CH.SC.U4CSE23240

---



```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    "Variable": [
        "Climate", "Housing", "Health", "Crime", "Transportation",
        "Education", "Arts", "Recreation", "Economy"
    ],
    1: [0.190, 0.544, 0.782, 0.365, 0.585, 0.394, 0.985, 0.520, 0.142],
    2: [0.017, 0.020, -0.605, 0.294, 0.085, -0.273, 0.126, 0.402, 0.150],
    3: [0.207, 0.204, 0.144, 0.585, 0.234, 0.027, -0.111, 0.519, 0.239],
}
data.keys()
```

```python
print(data['Variable'])
```

```python
df = pd.DataFrame(data)
features = df.drop(columns=["Variable"])
scaler = StandardScaler()
scaled_data = scaler.fit_transform(features)
pca = PCA(n_components=3)
x = pca.fit_transform(scaled_data)
print(x.shape)
```

```python
pca.components_
plt.figure(figsize=(8, 6))
plt.title("PCA 2D Projection")
plt.scatter(x[:, 0], x[:, 1], color="blue", s=100)
for i, var in enumerate(df["Variable"]):
    plt.text(x[i, 0] + 0.02, x[i, 1] + 0.02, var)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 8))
axis = fig.add_subplot(111, projection='3d')
axis.scatter(x[:, 0], x[:, 1], x[:, 2], color="blue", s=100)
for i, var in enumerate(df["Variable"]):
    axis.text(x[i, 0], x[i, 1], x[i, 2], var)
axis.set_xlabel("PC1")
axis.set_ylabel("PC2")
axis.set_zlabel("PC3")
plt.title("PCA 3D Projection")
plt.show()
```
